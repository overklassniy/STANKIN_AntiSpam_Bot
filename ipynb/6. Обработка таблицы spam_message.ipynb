## –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ø–∞–º-—Å–æ–æ–±—â–µ–Ω–∏–π

### –ß—Ç–µ–Ω–∏–µ —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ .json —Ñ–∞–π–ª–∞ —Ç–∞–±–ª–∏—Ü—ã spam_message

```python
with open("../data/spam_messages_02-09-2025.json", "r", encoding="utf-8") as f:
    json_data = json.load(f)
```

### –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ .json —Ñ–æ—Ä–º–∞—Ç–∞ –∫ pandas.DataFrame

```python
table = json_data["objects"][0]
```

```python
columns = [col["name"] for col in table["columns"]]
```

```python
rows = table["rows"]
```

```python
df = pd.DataFrame(rows, columns=columns)
```

```python
df['timedate'] = [datetime.fromtimestamp(timestamp).strftime("%d.%m.%Y %H:%M:%S") for timestamp in df['timestamp']]
```

```python
df['timedate'] = pd.to_datetime(df['timedate'], format='%d.%m.%Y %H:%M:%S')
```

```python
df
```

```python
df_filtered = df[df['timedate'] >= pd.to_datetime('20.07.2025 10:58:25', format='%d.%m.%Y %H:%M:%S')]
```

```python
df_filtered
```

### –°–æ–∑–¥–∞–Ω–∏–µ .json —Å–ø–∏—Å–∫–∞ —Å–ª–æ–≤–∞—Ä–µ–π —Å —Ç–µ–∫—Å—Ç–∞–º–∏ —Å–æ–æ–±—â–µ–Ω–∏–π –∏ –∫–ª–∞—Å—Å–æ–º 1

```python
processed_data = []
```

```python
for message in list(df_filtered["message_text"]):
    processed_entry = {
        "text": message,
        "label": 1
    }
    processed_data.append(processed_entry)
```

```python
with open('../data/processed_spam_messages_02-09-2025.json', 'w', encoding='utf-8') as outfile:
    json.dump(processed_data, outfile, ensure_ascii=False, indent=4)
```
